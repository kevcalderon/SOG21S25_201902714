```python
import pandas as pd

filename = list(uploaded.keys())[0]  # Obtener el nombre del archivo subido
df = pd.read_csv(filename, sep=";")  # Leer el archivo CSV
print(df.head())  # Mostrar las primeras filas
```

```python
resumen = df.describe()
print(resumen)
```

```python
correlacion = df.corr()

# Mostrar la matriz de correlación
print(correlacion)
```

```python
import matplotlib.pyplot as plt

# se tomará la columna de sulfur dioxide
columna = df.columns[5]


plt.figure(figsize=(10, 5))
df[columna].value_counts().sort_index().plot(kind='bar', color='skyblue')

# Configuración del gráfico
plt.title(f'Distribución de {columna}')
plt.xlabel(columna)
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns


columna = df.select_dtypes(include=['number']).columns[0]

plt.figure(figsize=(8, 5))
sns.histplot(df[columna], bins=20, kde=True, color='purple')


plt.title(f'Distribución de {columna}')
plt.xlabel(columna)
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación")
plt.show()
```
